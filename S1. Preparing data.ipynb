```python
import pandas as pd

df = pd.read_csv('odor_dataset_raw.csv')
print (len(df))
df.head()
```

```python
odor_counts = {}

for odors in df['descriptors']:
    for odor in odors.split(';'):
        if odor not in odor_counts:
            odor_counts[odor] = 1
        else:
            odor_counts[odor] += 1
len(odor_counts)
```

```python
odor_counts = {odor: count for odor, count in sorted(odor_counts.items(), key= lambda x: -x[1])}
odor_counts
```

```python
label_dict = {}
class_n = 0
for odor, count in odor_counts.items():
    label_dict[odor] = class_n
    class_n += 1
    if len(label_dict) == 100:
        break
    
label_dict
```

```python
all_x = []
all_y_odor = []
all_y_label = []

for x, y in zip(df['nonStereoSMILES'], df['descriptors']):
    for odor in y.split(';'):
        if odor in label_dict:
            all_x.append(x)
            all_y_odor.append(odor)
            all_y_label.append(label_dict[odor])
            break
len(all_x)
```

```python
split = ['train'] * 4000 + ['test'] * (len(all_x) - 4000)
simplified_df = pd.DataFrame({'SMILES': all_x, 'Odor': all_y_odor, 'Class': all_y_label, 'Split': split})
simplified_df.head()
```

```python
simplified_df.to_csv('odor_dataset.csv', index=None)
```
